```python
import pandas as pd
from chembl_webresource_client.new_client import new_client
import os

if not os.path.exists('data'):
    os.makedirs('data')
```

```python
molecule = new_client.activity
mol = molecule.filter(target_chembl_id='CHEMBL203', standard_type='IC50', standard_units='nM')
records = list(mol)

for i in records:
    del i['activity_properties']
    del i['ligand_efficiency']
    
```

```python
dataset = pd.DataFrame(records)
dataset.to_csv('data/egfr_raw.csv')

print(f"Shape: {dataset.shape}")
print(f"Saved to: data/egfr_raw.csv")
```
